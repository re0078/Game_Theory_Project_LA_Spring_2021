        <br>
        <font size='4'><b>برنامه ریزی خطی</b></font>
        <br>
        <font size='4'>
            <p align='justify'></p>
        مسائل برنامه ريزي خطي شامل بهينه سازي تابع هدفي خطي است که بايستي يکسري محدوديت در فرم هاي تساوي هاي خطي و نامساوي برقرار شوند.برای مثال فرض کنید برای دو متغیر
            $x1$ و $x2$ مقدار $x1 + x2$ را با داشتن فرض های زیر بیشینه کنیم :
            <br>
            $x1 + 2x2 \le 4$ <br>
            $4x1 + 2x2 \le 12$ <br>
            $-x1 + x2 \le 1$ <br>
            $x1 \ge 0 , x2\ge0$ <br>
            در واقع ما میخواهیم جمع دو عدد مثبت را بیشینه کنیم . خب اگر هیچ شرطی وجود داشت مقدار هر دو را مثبت بی نهایت قرار میدادیم تا این مقدار بیشینه شود . ولی تعدادی شرط داریم که اجازه ی این کار را نمیدهد . در واقع شرط ها فضای حالت ما را محدود میکنند . <br>
          میتوان به دید هندسی به این مسئله نگاه کرد.همانطور که در شکل مشخص است ، در واقع در کل صفحه ی مختصات ما به دنبال نقطه ای میگردیم که در ناحیه ی خاکستری رنگ باشد و مقدار $x1 + x2$ را بیشینه کند . کافی است خط های $x1 + x2 = c$ را رسم کنیم و بزرگنرین $c$ که این خط با خط ناحیه ی خاکستری برخورد دارد جواب مسئله ی ما میباشد .<br>

<font size='4'><b>ِDual</b></font>
    <br>
    <font size='4'>
    برای هر مسئله ی برنامه ریزی خطی یک dual تعریف میکنیم که برابر است با. dual یا همان دوگان را به این صورت تعریف میکنیم :
        <br>
        برای مسئله ی ماکسیمم سازی زیر :
        <br>
        $\text{maximize } c^Tx $ <br>
        $\text{subject to the constraint:} Ax\le b \text{ and } x\ge0$
        <br>
        دوگان را به این صورت تغریف میکنیم :
        <br>
        $\text{minimize } y^Tb $ <br>
        $\text{subject to the constraint:} y^TA\ge c^T \text{ and } y\ge0$
        <br>
        برعکسش هم برقرار است یعنی dual مسئله ی مینیمم سازی مسئبه ی ماکسیمم سازی می باشد.

<div style="direction:rtl;line-height:200%;">
<font size='4'>
    <p align='justify'></p>
    در واقع برای هر مسئله ی برنامه ریزی خطی ماکسیمم کردن تابع هدف dual آن یک مسئله ی مینیمم سازی تابع هدف است . 
    برای تمرین dual مسئله ی زیر را بیابید :
    <div style = "direction:ltl">
    $\text{maximize } x_1+x_2$
    <br>
    $\text{subject to :}$
    <br>
    $ x_1 + 2x_2 \le 4$
    <br>
    $4x_1 + 2x_2 \le 12$
    <br>
    $-x_1 + x_2 \le 1$

<div style="direction:rtl;line-height:200%;">
<font size='4'>
    <p align='justify'></p>
    پاسخ :
    <br>
    $\text{minimize } 4y_1+12y_2+y_3$

    $\text{subject to :}$
    <br>
    $ y_1 + 4y_2 - y_3 \ge 1$
    <br>
    $2y_1 + 2y_2 + y_3 \ge 1$

        <b> پیدا کردن تعادل نش با استفاده از برنامه ریزی خطی</b>
        <br>
        <font size='4'><b>بازی های zero sum</b></font>
        <br>
        <font size='4'>
        در این جا به بازی های zero sum  و پیدا کردن نقطه تعادل آن ها توسط برنامه ریزی خطی میپردازیم .
        طبق قضیه ی کمینه بیشینه میدانیم که در یک بازی دو نفره نقطه ی تعادل جایی است که نفر اول سعی کند مینیمم امتیازی که کسب میکند را ماکسیمم کند و یا اینکه معادلا نفر دوم سعی کند ماکسیمم امتیازی که نفر اول کسب میکند را مینیمم کند و می دانیم که تعادل بازی هم در همین نقطه اتفاق میفتد یعنی جایی که 2 بازیکن به این صورت بازی کنند.
            <br>
            یک بازی zero sum دو نفره در نظر بگیرید که
            $A_1$ مجموعه ی حرکت های نفر اول و $A_2$ مجموعه ی حرکت های نفر دوم باشد.
            اگر $<s_1^1,s_1^2, ... , s_1^n>$ استراتژی نفر اول باشد میدانیم که هدف این بازیکن این است که مینیمم امتیاز خود را بیشینه کند یا معادلا :

    درواقع همانطور که در بالا میبینید نفر اول به ازای هر حرکت نفر دوم مانند 
    $a_2^k$ برای امتیازی که نفر اول بدست می آورد یک کران پایین 
    $U_1^*$ قرار میدهیم و هدف ماکسیمم کردن این کران پایین است.
    <br>
    در مسئله ی بالا متغیر های ما 
    $s_1^1,s_1^2 , ... s_1^n,U_1^*$ هستند.
    بررسی کنید که چطور میتوان این مسئله ی بالا را به یک مسئله ی برتامه ریزی خطی تبدیل کرد؟
    $s_1^1,s_1^2 , ... s_1^n$ که از مسئله ی بالا بدست می آید در واقع همان استراتژی نفر اول در نقطه ی تعادل است.
    و $U_1^*$ همان امتیازی است که نفر اول کسب میکند .
    <br>
