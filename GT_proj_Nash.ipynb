بهترین گزینه برای متهم اول خیانت کردن است. به عبارت دیگر، استراتژی اعتراف اکیداً بر استراتژی سکوت
غلبه می کند. چون بازیکن اول عاقل است پس استراتژی سکوت را انتخاب نمی کند. از طرفی چون بازیکن دوم
می داند که بازیکن اول عاقل است پس استدلال می کند که بازیکن اول استراتژی سکوت را انتخاب نمی کند.
شرایط مشابهی برای بازیکن دوم حاکم است. لذا دو بازیکن عاقل در تعامل با یکدیگر در این بازی پروفایل (خیانت، خیانت)

                اگر فرض کنیم بازیکن iام استراتژی $s_i$ را انتخاب کرده استراتژی‌های دیگر بازیکنان را تعریف می‌کنیم
                $$s_{-i}=(s_1,...,s_{i-1},s_{i+1},...,s_n)$$
                لذا 
 \(s=(s_i,s_{-i})\) 
 است.  فرض کنید 
 \(s_i\)

 دو استراتژی بازیکن 
 \(i\)اُم باشد آنگاه:
 می‌گوییم استراتژی 
 \(s_i\)

 \(s'_i\) 
 اکیداً غالب است اگر به‌ازای هر استراتژی پروفایل رقبا 
 \(i\)، 
مانند 
 \(s_{-i}\)،

 \(s'_i\) 
 به‌طور ضعیف غالب است اگر به‌ازای هر استراتژی پروفایل برای رقبا 
 \(i\)، 
 یعنی 
 \(s_{-i}\)، 
 داشته باشیم:
\[U_i(s_i,s_{-i})\geq U_i(s'_i,s_{-i}).\]
 و همچنین برای حداقل یک استراتژی 
 \(s_{-i}\)، 
 داشته باشیم که 
\[U_i(s_i,s_{-i})\geq U_i(s'_i,s_{-i}).\] 
می‌گوییم که استراتژی 
 \(s_i\) 
 بر استراتژي 
 \(s'_i\) 
 به‌طور بسیار ضعیف غالب است اگر و تنها اگر به‌ازای هر استراتژی پروفایل برای رقبا 
 \(i\)، یعنی \(s_{-i}\)، 
 داشته باشیم:
 \[U_i(s_i,s_{-i})\geq U_i(s'_i,s_{-i}).\]
